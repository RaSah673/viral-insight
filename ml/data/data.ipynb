```python
search_response = youtube.search().list(
    type='video',
    q='coding',
    part='snippet',
    maxResults=50
).execute()

pprint.pprint(search_response)
```

```python
video_ids = [[item['id']['videoId'], item["snippet"]["channelTitle"], item["snippet"]["title"]] for item in search_response['items']]
video_ids
```

```python
statistics = youtube.videos().list(
    part='statistics',
    id=','.join(video_ids)
).execute()

statistics["items"][0]
```
